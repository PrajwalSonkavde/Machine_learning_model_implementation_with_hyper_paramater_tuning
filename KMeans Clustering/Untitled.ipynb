## KMeans

```python
from sklearn.cluster import KMeans
```

```python
km = KMeans(n_clusters=3)
clusters = km.fit_predict(df)
clusters
```

```python
df['cluster'] = clusters
```

```python
df['cluster'].value_counts()
```

```python
df.head()
```

```python
km.inertia_
```

```python
cluster0 = df[df['cluster']==0]
cluster1 = df[df['cluster']==1]
cluster2 = df[df['cluster']==2]
```

```python
for i in df.columns:
    for j in df.columns:
        plt.scatter(cluster0[i],cluster0[j],color='blue',label='cluster 0')
        plt.scatter(cluster1[i],cluster1[j],color='green',label='cluster 1')
        plt.scatter(cluster2[i],cluster2[j],color='red',label='cluster 2')
        plt.xlabel(i)
        plt.ylabel(j)
        plt.legend()
        plt.show()
```

```python
plt.scatter(cluster0['Annual Income ($)'],cluster0['Age'],color='blue')
plt.scatter(cluster1['Annual Income ($)'],cluster1['Age'],color='green')
plt.scatter(cluster2['Annual Income ($)'],cluster2['Age'],color='red')
```

# Using Elbow method (but here we want 3 cluster only so need to elbow method)

```python
sos = []
for i in range(2,20):
    km = KMeans(n_clusters=i)
    km.fit_predict(df.drop(['Outcome'],axis=1))
    sos.append(km.inertia_)
```

```python
plt.plot(range(2,20), sos, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Distortion')
plt.title('The Elbow Method using Distortion')
plt.show()
```
